### Background <a name="background"></a>

Food is a major men's need and the reason for new places opening where meals are provided. We go to restaurants to enjoy food, celebrate and have memorable moments. A majority of restaurateurs incorrectly assume that if a restaurant serves good quality food, that is enough to manage a profitable venture. Besides the quality contributions, you should focus your restaurant and turn it into a name that clients can trust. If you are unable to catch customers attention, chances are that your small venture will be closed. According to the Restaurant Startup & Growth magazine survey, 23% of the restaurants fail in the very first year. But the study also pointed out that if the restaurant survived the first year, then only 14% and 7% such restaurants close in the second and third year respectively. A recent study by Cornell University estimates the failure rate even higher. The study shows that 60% of restaurants are closed in the first year. This indicates that the opening phase is critical for restaurant startup.

### Problem <a name="problem"></a>

The most significant achievement for a new restaurant is to be successful and survive among the other food makers. To accomplish this task restaurateur must comprehend the market niche, target customers and competitors. Many unsuccessful entrepreneurs opened a restaurant without proper research of the market and customers. A market niche for our venture is Russian food restaurant.

### Interest <a name="interest"></a>

The task of that research is to explore target customers to know about their economic, social and other backgrounds, find out more about competitors and choose the best place for opening in Toronto.

## Data <a name="data"></a>

Based on definition of our problem, factors that will influence our decission are:
* neighbourhoods with the most Russian language speakers
* economic background of potential customers
* number of existing Russian/Ukraine/Polish restaurants in Toronto
* distance of neighborhood from city center

Following data sources will be needed to extract/generate the required information:
* Toronto demographic data will be generated using **Statistic Canada** site. 
* centers of candidate areas will be generated algorithmically and approximate addresses of centers of those areas will be obtained using **Foursquare API**
* number of restaurants and their type and location in every neighborhood will be obtained using **Foursquare API**
* coordinate of Toronto center will be obtained using CSV file scraped from **Wikipedia**
